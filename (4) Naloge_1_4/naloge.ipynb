```python
import re
import sys
print(sys.version)
```

```python
# Poisce vse besede, ki so zacetek stavka, ali pa sledijo besedi ”and” - uporaba
# (? <= ...)

poisci = re.compile(r'\b(?:(?<=and)\s|^|\.\s)([a-z]+)', re.IGNORECASE)
data = poisci.findall('Buying chesse AND salami AND wine AND bread. Soup AND veggies.')
print(data)
```

```python
# Poisce vse besede, ki se zacnejo s samoglasnikom in se koncajo s soglasnikom -
# uporaba [abcd] in [ˆabcd]
poisci = re.compile(r'\b[aeiou][a-z]*[^aeiou.\s]\b', re.IGNORECASE)
data = poisci.findall('ABC ded aba UEB AOB. aaa ded aba UEB aoa.')
print(data)
```

```python
#Poisce vse stringe, ki so znotraj oklepajev (ampak brez samih oklepajev) -
# uporaba (? <= ...) in (? = ...) in non-greedy quantifiers
poisci = re.compile(r'\(([^)]+)\)')
data = poisci.findall('Tezka naloga (3.naloga) tezka ampak ni tako tezka kot (prva) naloga.')
print(data)
```

```python
# 3. naloga druga moznost

poisci = re.compile(r'(?<=\()[^()]+(?=\))')
data = poisci.findall('Tezka naloga (3.naloga) tezka ampak ni tako tezka kot (prva) naloga.')
print(data)
```

```python
# Poisce vsa petmestna stevila, ki se nazaj in naprej preberejo enako (npr. 12321) -
# uporaba \1
poisci = re.compile(r'(\d)(\d)(\d)(\2)(\1)')
# x y z (\2)..enaka 2. stevilki torej y, \1 pa enaka prvi torej x
data = poisci.findall('12321 balsadas dsaa  111111 spijdsodsdsoi98789 as21234')
print(data)
```

```python
# ALI 4. nal
p = re.compile(r'(\d)(\d)(\d)\2\1')
text = "12321 balsadas dsaa  111111 spijdsodsdsoi98789 as21234"
for m in p.finditer(text):
    print(m.group(0))   # celotno ujemanje
```

```python
#  Poisce vse trimestne stringe, ki ne sledijo stevilu in jim ne sledi stevilo - uporaba
# (? <!...) in (?!...)
poisci = re.compile(r'(?<!\d)[a-z]{3}(?!\d)', re.IGNORECASE)
data = poisci.findall('12321balsadasdsaa11111spijdsodsdsoi98789 as21234 ')
print(data)
```
